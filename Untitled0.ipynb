```python
data = [35, 20, 25, 5 ,15]
lb=['Kerry', 'Flash', 'j&T', 'DHL','Thailand Post']
plt.pie(data,labels=lb, autopct="%.0f%%", colors=['red','yellow','blue','lightgreen','green'])
plt.show()
```

```python
data = [35, 20, 25, 5 ,15]
lb=['Kerry', 'Flash', 'j&T', 'DHL','Thailand Post']
explode = [0, 0.1, 0 ,0.2, 0]
plt.pie(data, labels=lb, explode=explode,startangle =90, colors=['red','yellow','blue','lightgreen','green'] ,shadow=True,autopct="%.1f%%")
plt.show()
```
